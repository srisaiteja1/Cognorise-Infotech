```python
import pandas as pd
```

```python
house_data=pd.read_csv("data.csv")
house_data
```

```python
house_data = house_data[['price', 'bedrooms','bathrooms', 'sqft_living', 'floors']]
house_data.head()
```

```python
cutoff = int(len(house_data) * 0.8)  # 4600 * 0.8

house_data_train = house_data[:cutoff]


house_data_test = house_data[cutoff:]
```

```python
X_train = house_data_train[['bedrooms', 'bathrooms', 'sqft_living', 'floors']]
X_test = house_data_test[['bedrooms', 'bathrooms', 'sqft_living', 'floors']]

# our output
y_train = house_data_train['price']
y_test = house_data_test['price']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()  
model.fit(X_train, y_train)  
```

```python
m = model.coef_
m
```

```python
c = model.intercept_
c
```

```python
equation = ""
for i, column in enumerate(X_train.columns):
    equation = f"({m[i]} x {column})" + " + " + equation

equation = equation + f"{c}"

print("Our Equation is : \n", equation)
```

```python
y_pred = model.predict(X_test)
```

```python
print(f"True price is: {list(y_test)[1]} a predicted price is {y_pred[1]}")
print(f"True price is: {list(y_test)[2]} a predicted price is {y_pred[2]}")
print(f"True price is: {list(y_test)[3]} a predicted price is {y_pred[3]}")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

baseline_mae = mean_absolute_error([y_train.mean()] * len(y_train), y_train)

pred_mae = mean_absolute_error(y_test, y_pred)
pred_mse = mean_squared_error(y_test, y_pred)
pred_mae = mean_absolute_error(y_test, y_pred)

print(f'baseline {baseline_mae}')
print(f'baseline {pred_mae}')
print(f'baseline {pred_mse}')
```
